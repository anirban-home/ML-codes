# 7. Outlier detection by IQR

```python
import numpy as np
df.sort_values('price',inplace=True,ignore_index=True)
Q1=np.percentile(df['price'], 25)
Q2=np.percentile(df['price'], 50)
Q3=np.percentile(df['price'], 75)
IQR=Q3-Q1
out_low=Q1-(1.5*IQR)
out_hi=Q3+(1.5*IQR)
outlier=[]
for i in range(205):
    if df.iloc[i,25]>out_hi or df.iloc[i,25]<out_low:
        outlier.append(df.iloc[i,25])
print('The outliers are:')
for i in range(len(outlier)):
    print(outlier[i])
    
```

# 8. Calculate variance

```python
mean_price=df['price'].mean()
#var_price=df['price'].var()
var_price=0
for i in range (205):
     var_price= var_price+((df.iloc[i,25]-mean_price)**2)
var_price=var_price/204    
print('Variance of price  :',var_price)
```

# 9. Calculate standard deviation

```python
import math
#sd_price=df['price'].std()
sd_price=math.sqrt(var_price)
print('Standard Deviation of price: ',sd_price)
```
